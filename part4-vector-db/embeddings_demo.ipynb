```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sentences = [
    # Cricket
    "The batsman scored a century in the match",
    "The bowler delivered a fast yorker",
    "The team won the cricket tournament",
    "The player hit a six over the boundary",

    # Cooking
    "The chef prepared a delicious pasta",
    "Boil the rice until it is soft",
    "Add spices to enhance the flavor",

    # Cybersecurity
    "The system was attacked by a hacker",
    "Use strong passwords to secure accounts",
    "Firewall protects the network from threats"
]
```

```python
sentences = [
    # Cricket
    "The batsman scored a century in the match",
    "The bowler delivered a fast yorker",
    "The team won the cricket tournament",
    "The player hit a six over the boundary",

    # Cooking
    "The chef prepared a delicious pasta",
    "Boil the rice until it is soft",
    "Add spices to enhance the flavor",

    # Cybersecurity
    "The system was attacked by a hacker",
    "Use strong passwords to secure accounts",
    "Firewall protects the network from threats"
]
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)
```

```python
similarity_matrix = cosine_similarity(embeddings, embeddings)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(similarity_matrix, annot=True, cmap='coolwarm')
plt.title("Cosine Similarity Matrix")
plt.show()
```

```python
query = "The bowler took three wickets in one over"
query_embedding = model.encode([query])

similarities = cosine_similarity(query_embedding, embeddings)[0]

# Get top 2 similar sentences
top_indices = similarities.argsort()[-2:][::-1]

for idx in top_indices:
    print("Sentence:", sentences[idx])
    print("Similarity Score:", similarities[idx])
    print()
```
